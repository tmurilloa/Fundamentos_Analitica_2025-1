```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../files/input/solicitudes_de_credito.csv", sep=";")
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
def limpiar_texto(cadena):
    if isinstance(cadena, str):
        cadena = cadena.lower()           # a minúsculas
        cadena = cadena.strip()
        cadena = cadena.replace(' ', '_')  # quitar espacios
        cadena = cadena.replace('-', '_') # reemplazar '-' por '_'
    return cadena

df_limpio = df.applymap(limpiar_texto)
```

```python
df_limpio.head(24)
```

```python
from datetime import datetime

def parsear_fecha(fecha_str):
    for formato in ("%d/%m/%Y", "%Y/%m/%d"):
        try:
            return datetime.strptime(fecha_str, formato)
        except ValueError:
            continue
    return pd.NaT  # si no pudo interpretar el formato

df_limpio['fecha_de_beneficio'] = df_limpio['fecha_de_beneficio'].apply(parsear_fecha)
```

```python
df_limpio.head(30)
```

```python
df_limpio['monto_del_credito'] = df_limpio['monto_del_credito'].astype(str).str.replace(r'[\$,._\s]', '', regex=True)
df_limpio['monto_del_credito'] = pd.to_numeric(df_limpio['monto_del_credito'], errors='coerce')
```

```python
df_limpio.head(25)
```

```python
df_limpio['comuna_ciudadano'] = pd.to_numeric(df_limpio['comuna_ciudadano'], errors='coerce')
```

```python
df_limpio.head(25)
```

```python
df_limpio.shape
```

```python
df_sin_duplicados = df_limpio.drop_duplicates()
df_sin_duplicados.shape
```

```python
df_sin_duplicados.head(25)
```

```python
(df_sin_duplicados == 'nan').sum()
```

```python
df_sin_duplicados = df_sin_duplicados.dropna()
```

```python
df_sin_duplicados.to_csv("../files/output/solicitudes_de_credito.csv", index=False, sep=';')
```
